### 1.4 Tratamento dos dados

```python
df['normalized-losses'] = pd.to_numeric(df['normalized-losses'],errors='coerce')
df['stroke'] = pd.to_numeric(df['stroke'],errors='coerce')
df['bore'] = pd.to_numeric(df['bore'],errors='coerce')
df['horsepower'] = pd.to_numeric(df['horsepower'],errors='coerce')
df['peak-rpm'] = pd.to_numeric(df['peak-rpm'],errors='coerce')
df['price'] = pd.to_numeric(df['price'],errors='coerce')
```

```python
df = df.sort_values(by='price', ascending=False)
df.head()
```

```python
df.info()
```

Após os ajustes é possível observar que a variável 'normalized-losses' possui 20% dos seus dados nulos, sendo a variável com a maior quantidade de dados faltantes. Variáveis que também apresentaram dados faltantes foram: bore, stroke, horsepower, peak-rpm e price.

## 2.1 Visualizações

Iniciarei verificando as relações existentes entre as variáveis númericas e a variável alvo.

```python
correlacao = df.corr()
f, co = plt.subplots(figsize = (15,9))
sns.heatmap(correlacao, annot=True);
```

Com o heatmap atrelado a função corr fica fácil de identificar quais variáveis numéricas tem maior correlação com a variável price. Logo abaixo é possível observar algumas das variáveis com maior índice de correlação.

```python
fig = plt.figure(figsize=(15,15))

graf1 = fig.add_subplot(321)
graf2 = fig.add_subplot(322)
graf3 = fig.add_subplot(323)
graf4 = fig.add_subplot(324)
graf5 = fig.add_subplot(325)
graf6 = fig.add_subplot(326)

graf1.scatter(df['length'], df['price'])
graf2.scatter(df['width'], df['price'])
graf3.scatter(df['curb-weight'], df['price'])
graf4.scatter(df['engine-size'], df['price'])
graf5.scatter(df['horsepower'], df['price'])
graf6.scatter(df['highway-mpg'], df['price'])

graf1.set_title('length x price')
graf1.set_xlabel('length')
graf1.set_ylabel('price')

graf2.set_title('width x price')
graf2.set_xlabel('width')
graf2.set_ylabel('price')

graf3.set_title('curb-weight x price')
graf3.set_xlabel('curb-weight')
graf3.set_ylabel('price')

graf4.set_title('engine-size x price')
graf4.set_xlabel('engine-size')
graf4.set_ylabel('price')

graf5.set_title('horsepower x price')
graf5.set_xlabel('horsepower')
graf5.set_ylabel('price')

graf6.set_title('highway-mpg x price')
graf6.set_xlabel('highway-mpg')
graf6.set_ylabel('price')

fig.tight_layout()
plt.show()
```

Conforme observado pelo gráfico anterior as variáveis length, width, curb-weight, engine-size e horsepower apresentam uma forte correlação positiva já a variável highway-mpg apresenta um forte correlação negativa (OBS: A variável city-mpg não foi descatada por ser semelhante à highway-mpg).

```python
plt.figure(figsize=(15, 6)) 
sns.displot(data=df, x="engine-size", hue="engine-type", col="engine-type")
plt.xlabel('Tamanho do motor') 
```

No gráfico acima faço um comparação entre as variáveis tamanho do motor e o tipo do motor, considerando que o tamanho possui uma forte correlação positiva com o preço é possível afirmar que motores maiores agregam um custo maior aos veículos. Pelo gráfico temos muitos dados relacionados ao tipo de motor ohc que possuem em sua maioria um tamanho que vai de 70 a 170 unidades, ou seja, apesar da quantidade quando comparado aos outros tipos de motor este possui um dos menores tamanhos, sendo superior apenas ao do tipo rotor. Isto indicando que seu preço médio deve ser menor quando comparado a maioria dos outros modelos. O motor do tipo ohcv se destaca por possuir os maiores modelos dentre os dados, o que pode corresponder também a maior média de preço.

```python
df_ohcv = df.loc[df['engine-type'] == 'ohcv']
df_ohc = df.loc[df['engine-type'] == 'ohc']
df_ohcf = df.loc[df['engine-type'] == 'ohcf']
df_dohc = df.loc[df['engine-type'] == 'dohc']
df_l = df.loc[df['engine-type'] == 'l']
df_rotor = df.loc[df['engine-type'] == 'rotor']
df_dohcv = df.loc[df['engine-type'] == 'dohcv']

print(df_ohcv['price'].mean())
print(df_ohc['price'].mean())
print(df_ohcf['price'].mean())
print(df_dohc['price'].mean())
print(df_l['price'].mean())
print(df_rotor['price'].mean())
print(df_dohcv['price'].mean())
```

Observando a média de preço dos veículos com cada tipo de motor podemos afirmar que os motores do tipo ohcv estão no geral presente em veículos com maiores preços e o motor do tipo ohc em veículos com menores preços. Isto pode estar atrelado também ao fabricante e a tecnologia empregada em cada tipo destes motores.

```python
plt.figure(figsize=(14,7))
#sns.set(rc={'xtick.labelsize':17,'ytick.labelsize':10,'axes.labelsize':15 , "axes.grid":True})
sns.set_style("whitegrid")

# Use "Pallete" to specify the colors to be used for different levels of the hue
sns.lineplot(x="price" , y="engine-size" , hue = "num-of-cylinders" , dashes = False,palette = 'tab10' ,data=df)

plt.xlabel('Preço')  
plt.ylabel('Tamanho do Motor')  
plt.title('Consumo x Potência para cada tipo de tração') 


plt.show()
```

Através do gráfico anterior é possível obsservar uma tendência dos carro com uma maior quantidade de cilindros possuirem um motor maior como também um preço maior.

```python
plt.figure(figsize=(14,7))
#sns.set(rc={'xtick.labelsize':17,'ytick.labelsize':10,'axes.labelsize':15 , "axes.grid":True})
sns.set_style("whitegrid")

# Use "Pallete" to specify the colors to be used for different levels of the hue
sns.lineplot(x="price" , y="width", hue = "fuel-system" , dashes = False,palette = 'tab10' ,data=df)

plt.xlabel('Preço')  
plt.ylabel('Largura do veículo')  
plt.title('Relação entre o tipo de sistema de injeção com a largura do veículo e o preço') 


plt.show()
```

No gráfico é analisado como o tipo de sistema de injeção se relaciona com as variáveis preço e largura do veículo. Dentre os pontos interessantes para se destacar tem o tipo 2bbl que possue um valor máximo ligeiramente acima dos 10000 o que pode indicar ser um tipo de sistema de injeção presente em carros populares cujo preço deve estar dentre as menores faixas de valores, o tipo mpfi que possui a maior abrangência dentre os automóveis deste conjunto de dados podendo assim estar presente em todo tipo de modelo e a tipo idi que possui uma presença semelhante ao tipo mpfi. No geral apesar do gráfico exibir uma tendência para tipos de diferentes de sistema de injeção ele também apresenta esses sistemas possuindo um variação de preço semelhante em suas respectivas faixas indicando que está variável pode não ter uma correlação forte com o preço.

```python
plt.figure(figsize=(14,7))
sns.set_style("whitegrid")

sns.lineplot(x="horsepower" , y="city-mpg" , hue = "drive-wheels" , dashes = False,palette = 'CMRmap' ,data=df)

plt.xlabel('Potência do Motor (HP)')  
plt.ylabel('Milhas por galão de combustível consumido ( Cidade )')  
plt.title('Consumo x Potência para cada tipo de tração') 


plt.show()
```

Através deste gráfico é possível evidenciar a relação entre as variáveis potência do motor, consumo de combustível na cidade e o tipo de tração. Quanto maior a potêcia do motor menor a quantidade de milhas por galão de combustível, carros com potência acima de 150HP apresenta com uma maior frequência a tração do tipo traseira (rwd).

```python
plt.figure(figsize=(14,7))

sns.lineplot(x="horsepower" , y="highway-mpg" , hue = "drive-wheels" , dashes = False,palette = 'CMRmap' ,data=df)

plt.xlabel('Potência do Motor (HP)')  
plt.ylabel(' Milhas por galão de combustível consumido ( Estrada )')  
plt.title('Consumo x Potência para cada tipo de tração') 

plt.show()
```

Para fins de comparação com o gráfico que utiliza o consumo dos veículos na cidade este outro gráfico exibe a avalição com utilizando o consumo na estrada. Desta forma é possível notar uma disposição semelhante dos dados portanto, não há grande discrepância entre as duas análises.

```python
plt.figure(figsize=(10, 4))
sns.boxplot(data=df, x="price", y="body-style",width=0.7, color='cornflowerblue')
plt.xlabel('Preço') 
plt.ylabel('Categoria') 
plt.title('Distribuição dos preços dos carros para cada categoria')
plt.show()
```

O que é possível inferir do gráfico é que cada categoria de carro possui uma faixa de preço distinta sendo o hardtop o que possui os carros com preços mais elevados e o hatchback a categoria que possui os carros com preços menores.
